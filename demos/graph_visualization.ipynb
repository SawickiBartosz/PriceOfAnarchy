```python
import demo_fix
from dtos import q, Graph, Flow

G = Graph('A', 'D', [
    ('A', 'B', q(0,1,0)),
    ('A', 'C', q(0,0,10)),
    ('C', 'B', q(0,1,0)),
    ('B', 'D', q(0,0,10)),
    ('C', 'D', q(0,1,0))
])
G.draw()
print(G.possible_paths())
```

```python
flow = Flow(G, {
    ('A', 'C', 'B', 'D'): 1, 
    ('A', 'B', 'C', 'D'): 2, 
    ('A', 'B', 'D'): 3, 
    ('A', 'C', 'D'): 4
})
flow.draw(simple=True); 
flow.draw(); 
```
