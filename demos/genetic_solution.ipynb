```python
import demo_fix
from dtos import q, Graph
from solutions import GeneticSolution, TargetOptimal
from common import DiscreteCore, ContinuousCore
from visualization import Plotter

G = Graph('A', 'D', [
    ('A', 'B', q(0,1,0)),
    ('A', 'C', q(0,0,10)),
    ('C', 'B', q(0,1,0)),
    ('B', 'D', q(0,0,10)),
    ('C', 'D', q(0,1,0))
])
solution = GeneticSolution(G).solve(DiscreteCore(10), TargetOptimal())
solution.draw()
```

```python
solution = GeneticSolution(G).solve(ContinuousCore(10), TargetOptimal(), seed=13)
solution.draw()
```

```python
solution = GeneticSolution(G, num_generations=10, sol_per_pop=10).solve(ContinuousCore(10), TargetOptimal(), attach=Plotter(), seed=13)
solution.draw()
```
