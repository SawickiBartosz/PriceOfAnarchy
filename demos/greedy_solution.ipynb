```python
import demo_fix
from dtos import q, Graph, Flow
from solutions import GreedySolution, TargetOptimal, TargetNE
from common import DiscreteCore

G = Graph('A', 'D', [
    ('A', 'B', q(0,1,0)),
    ('A', 'C', q(0,0,10)),
    ('C', 'B', q(0,1,0)),
    ('B', 'D', q(0,0,10)),
    ('C', 'D', q(0,1,0))
])
solution = GreedySolution(G).solve(DiscreteCore(10), TargetOptimal())
solution.draw()
```

```python
solution = GreedySolution(G).solve(DiscreteCore(10, precision=0.01), TargetOptimal())
solution.draw()
```
