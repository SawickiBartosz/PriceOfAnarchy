```python
import demo_fix
from dtos import q, poly, Graph
from solutions import GreedySolution, RotatingGreedySolution, GeneticSolution, TargetOptimal, TargetNE
from common import DiscreteCore, ContinuousCore
from utils import Context

greedy_methods = Context({
    "No rotations": GreedySolution,
    "1 rotation": RotatingGreedySolution,
    "5 rotations": RotatingGreedySolution.parametrize(rotations=5),
})

G = Graph('A', 'D', [
    ('A', 'B', q(1,0,0)),
    ('A', 'C', q(0,0,10)),
    ('C', 'B', poly(0,3,0,0,0)),
    ('B', 'D', q(0,0,10)),
    ('C', 'D', q(0,1,0))
])
greedy_methods.run_and_draw(G, DiscreteCore(10), TargetOptimal())
greedy_methods.run_and_draw(G, DiscreteCore(10), TargetNE())
```

```python
greedy_methods.run_and_draw(G, DiscreteCore(10, precision=1/128), TargetOptimal())
greedy_methods.run_and_draw(G, DiscreteCore(10, precision=1/128), TargetNE())
```

```python
genetic_methods = Context({
    "100 generation, 100 per generation": GeneticSolution.parametrize(num_generations=100, sol_per_pop=100),
    "10 generation, 1000 per generation": GeneticSolution.parametrize(num_generations=10, sol_per_pop=1000),
    "1000 generation, 10 per generation": GeneticSolution.parametrize(num_generations=1000, sol_per_pop=10),
})
genetic_methods.run_and_draw(G, ContinuousCore(10), TargetOptimal())
```
